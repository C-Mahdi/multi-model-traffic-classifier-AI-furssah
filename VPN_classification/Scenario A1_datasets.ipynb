```python
from scipy.io import arff
import pandas as pd

# Load ARFF file using scipy
data, meta = arff.loadarff('/dataset/TimeBasedFeatures-Dataset-15s-VPN.arff')

# Convert to pandas DataFrame
df = pd.DataFrame(data)

# Decode byte strings (especially in the 'class1' or any categorical column)
for col in df.select_dtypes(['object']).columns:
    df[col] = df[col].str.decode('utf-8')

# Display first rows
df.head()

```

```python
df.to_csv('TimeBasedFeatures-Dataset-15s-VPN.csv', index=False)
print("CSV saved successfully.")
```

```python
import pandas as pd

# Read the ARFF file as raw text (manual approach)
with open('TimeBasedFeatures-Dataset-60s-VPN.arff', 'r') as f:
    lines = f.readlines()

# Find the start of data section
data_start = None
for i, line in enumerate(lines):
    if '@data' in line.lower():
        data_start = i + 1
        break

# Extract attribute lines and data lines
attribute_lines = lines[:data_start]
data_lines = lines[data_start:]

# Get attribute names
attributes = []
for line in attribute_lines:
    if line.lower().startswith('@attribute'):
        attr_name = line.split()[1].strip("'\"")
        attributes.append(attr_name)

# Create a clean list of data (replace '' or '?' with NaN)
clean_data = []
for line in data_lines:
    if line.strip() == '':
        continue  # skip empty lines
    line = line.strip().replace('?', 'nan').replace(',,', ',nan,')  # optional double commas
    # Ensure lines ending with comma get a NaN
    if line.endswith(','):
        line += 'nan'
    split_line = [x.strip() if x.strip() != '' else 'nan' for x in line.split(',')]
    clean_data.append(split_line)

# Create DataFrame
df2 = pd.DataFrame(clean_data, columns=attributes)

# Convert numeric columns
for col in df2.columns:
    df2[col] = pd.to_numeric(df2[col], errors='ignore')

df2
```

```python
# Count the occurrences of each class
print(df2['class1'].value_counts())
```

```python
df1.to_csv('TimeBasedFeatures-Dataset-60s-VPN.csv', index=False)
print("CSV saved successfully.")
```

```python
df_15s = pd.read_csv('TimeBasedFeatures-Dataset-15s-VPN.csv')
df_30s = pd.read_csv('TimeBasedFeatures-Dataset-30s-VPN.csv')
df_60s = pd.read_csv('TimeBasedFeatures-Dataset-60s-VPN.csv')
df_120s = pd.read_csv('TimeBasedFeatures-Dataset-120s-VPN.csv')
```

```python
common_cols = set(df_15s.columns) & set(df_30s.columns) & set(df_60s.columns)
df_15s = df_15s[list(common_cols)]
df_30s = df_30s[list(common_cols)]
df_60s = df_60s[list(common_cols)]
df_120s = df_120s[list(common_cols)]
```

```python
# Concatenate them vertically
df_combined = pd.concat([df_15s, df_30s, df_60s, df_120s], ignore_index=True)

# Optional: shuffle the combined dataset
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_combined.to_csv('Combined_TimeBasedFeatures_Dataset.csv', index=False)
print("All datasets combined and saved as Combined_TimeBasedFeatures_Dataset.csv")
```
